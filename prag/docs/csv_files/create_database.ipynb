```python
import pandas as pd
import os
```

```python
policy = pd.read_csv('./policy_info_topic.csv',encoding='utf-8')
policy.info()
```

```python
policy_1 = policy[policy['couny'] == '国内']
policy_1 = policy_1[policy_1['contentText'].isnull() == False]

policy_1.head()
```

```python
"""
将title，channel，contentText列中的文本保存在./raw_data中。
格式为：id_title.txt，内容为：
# -*- coding: utf-8 -*-

## 政策标题：
政策标题

## 政策来源：
政策来源

## 政策内容：
政策内容

同时，生成一张表格，包含id，title，file_name，file_path等信息。

"""


def clean_title(title):
    # clean title by 转义字符 like '\r'
    chr = ['\r', '\n', '\t', '\\', '/', ':', '*', '?', '"', '<', '>', '|', ' ','。', '，', '。', '！', '；', '：', '“', '”', '‘', '’']
    for c in chr:
        title = title.replace(c, '')
    return title

fid = 0
for index, row in policy_1.iterrows():
    title = row['title']
    title = clean_title(title)
    channel = row['channel']
    contentText = row['contentText']
    id = fid
    fid += 1

    file_name = str(id) + '_' + title + '.txt'
    file_path = os.path.join('..//raw_data', file_name)

    policy_1.at[index, 'file_name'] = file_name
    policy_1.at[index, 'file_path'] = file_path
    policy_1.at[index, 'id'] = id

    with open(file_path, 'w', encoding='utf-8') as f:
        f.write('# -*- coding: utf-8 -*-\n\n')
        f.write('## 政策标题：\n')
        f.write(title + '\n\n')
        f.write('## 政策来源：\n')
        f.write(channel + '\n\n')
        f.write('## 政策内容：\n')
        f.write(contentText + '\n')



policy_1.head()
```
